```python
# Remove duplicates
df.drop_duplicates(inplace=True)

```

```python
# Fill or drop missing values
df.dropna(inplace=True)

```

```python
# Fix date column
df['Order Date'] = pd.to_datetime(df['Order Date'])

```

```python
# Extract Year and Month
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
# Total Sales
print(df['Sales'].sum())

```

```python
# Top 10 Products
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
#YoY Growth
yearly = df.groupby('Year')['Sales'].sum().reset_index()
yearly['YoY_Growth'] = yearly['Sales'].pct_change() * 100
print(yearly)
```

```python
#West Region Anomaly:
west = df[df['Region'] == 'West']
west_analysis = west.groupby('Category')[['Sales','Profit']].sum()
west_analysis['Margin'] = (west_analysis['Profit'] / west_analysis['Sales']) * 100
print(west_analysis)
```
